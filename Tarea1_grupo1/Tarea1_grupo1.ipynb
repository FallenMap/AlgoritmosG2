```python
# Python program to demonstrate searching operation
# in binary search tree without recursion
class newNode:

		# Constructor to create a new node
		def __init__(self, data):
				self.data = data
				self.left = None
				self.right = None

# Function to check the given
# key exist or not
def iterativeSearch(root, key):
		
		# Traverse until root reaches
		# to dead end
		while root != None:
				
				# pass right subtree as new tree
				if key > root.data:
						print(chr(27)+"[31m"+"Nodo: "+str(root.data),end="") 
						print(chr(27)+"[37m"+"--> Arriba -->",end="")
						root = root.right
						

				# pass left subtree as new tree
				elif key < root.data:
						print(chr(27)+"[31m"+"Nodo: "+str(root.data),end="") 
						print(chr(27)+"[37m"+"--> Abajo -->",end="")
						root = root.left
						
				else:
						print(chr(27)+"[32m"+"Nodo: "+str(root.data),end="") 
						print(chr(27)+"[37m"+"--> FIN de la busqueda",end="")
						return True # if the key is found return 1
		return False

# A utility function to insert a
# new Node with given key in BST
def insert(Node, data):
		
		# If the tree is empty, return
		# a new Node
		if Node == None:
				return newNode(data)

		# Otherwise, recur down the tree
		if data < Node.data:
				Node.left = insert(Node.left, data)
		elif data > Node.data:
				Node.right = insert(Node.right, data)

		# return the (unchanged) Node pointer
		return Node

def custom_console(root):

	# Get tree nodes count
	cantidad_nodos = int( input("Ingrese la cantidad de nodos que conforman el arbol sin contar la raiz: ") )
 
 # Get each single node value from console input
	for dummy_id in range( cantidad_nodos):
		nodo_value = input("Valor ( "+str(dummy_id+1)+"): ")
		root = insert( root, nodo_value) # Update tree's root node
		

def printBinaryTree(root, space, height): # O(n)
		# It is necesary to iterate over all nodes in tree, regardless it's a 
		# recursive algorithm. Also note that there is not a constant factor
		# which specifies how many nodes are in a subtree after a recursive
		# call (since tree can be unbalanced)
 
    # Base case
    if root is None:
        return

    # increase distance between levels
    space += height
 
    # print right child first
    printBinaryTree(root.right, space, height)
    print()
 
    # print the current node after padding with spaces
    for i in range(height, space):
        print(' ', end='')
 
    print(root.data, end='')
 
    # print left child
    print()
    printBinaryTree(root.left, space, height)



# Driver Code
if __name__ == '__main__':
		
		# Let us create following BST
		#    50
		#  30   70
		# / \  / \
		#20 40 60 80
		"""root = None
		root = insert(root, 50)
		insert(root, 30)
		insert(root, 20)
		insert(root, 40)
		insert(root, 70)
		insert(root, 60)
		insert(root, 80)"""

		# Inits tree
		root=None

		# Gets and sets tree root's value
		root_value = input("Ingrese el valor de la raiz: ")
		root = insert( root, root_value)
	
		# rootAux will reference tree root node
		rootAux=root

		# Build the tree by means of user inputs
		custom_console(root)

		# Gets a value to search over the tree
		valor_a_buscar = input("Ingrese el valor a buscar en el arbol: ")
	
		print("\nNOTA: La rama del arbol en la parte superior representa su rama derecha\n")
		# Prints tree recursively
		printBinaryTree(rootAux,0,10)
		print("")

		# Search for a value in the tree
		if iterativeSearch(rootAux, valor_a_buscar):
			print()
		else:
				print("")
				print("El valor de busqueda NO fue encontrado")
```
